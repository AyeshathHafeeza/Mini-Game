```python
# Number Guessing using Python
import random as re
secret_num=re.randint(1,100)

for i in range(0,secret_num+1):
  guess=int(input("Guess number:"))
  if guess>secret_num:
    print("Too High, Try Again")
  elif guess<secret_num:
    print("Too Low, Try Again")
  else:
    print("Correct Guess")
    break
```
